```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data.csv')
```

```python
df.head()
```

```python
def extract_price(price):
    if 'грн.' in price:
        return int(''.join(filter(str.isdigit, price)))
    else:
        return None

df['Ціна'] = df['Ціна'].apply(extract_price)    
    
```

```python
df.describe()
```

```python
def split_city_district(city):
    city_parts = city.split(', ')
    if len(city_parts) == 2:
        return city_parts[0], city_parts[1]
    else:
        return city_parts[0], None
    
df[['Місто', 'Район']] = df['Місто'].apply(split_city_district).apply(pd.Series)
df
```

```python
df['Місто'].nunique()
```

```python
df.groupby('Місто').count()
```

```python
df['Загальна площа'].describe()
```

```python
# Функція для визначення кількості кімнат на основі площі
def determine_qty_room(area):
    if area < 50:
        return '1-кімнатна'
    elif 50 <= area < 70:
        return '2-кімнатна'
    elif 70 <= area < 90:
        return '3-кімнатна'
    else:
        return 'n-кімнатна'

# Додавання нового стовпця 'qty room'
df['qty room'] = df['Загальна площа'].apply(determine_qty_room)
```

```python
df['qty room'].value_counts()
```

```python
grouped_counts = df.groupby(['Місто', 'ID оголошення']).size().reset_index(name='Кількість')
grouped_counts
```

```python
df['qty room'].value_counts().plot.pie(autopct=lambda p:f'{p:.0f}%',
                                  colors=sns.color_palette('Set3'),
                                  shadow=True)
```

```python
# Підрахунок кількості оголошень по кожному місту
city_counts = df['Місто'].value_counts()

# Візуалізація стовпчастою діаграмою з підписами
plt.figure(figsize=(12, 6))
bars = plt.bar(city_counts.index, city_counts.values, color='skyblue')

# Додавання підписів до кожного стовпця
for bar, count in zip(bars, city_counts.values):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 0.1, str(count), ha='center', color='black')

plt.title('Кількість оголошень по містах')
plt.xlabel('Місто')
plt.ylabel('Кількість оголошень')
plt.xticks(rotation=45, ha='right')
plt.show()
```
